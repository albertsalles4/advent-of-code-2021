```python
def preprocess_input(input_list):
    preprocess = []
    for string in input_list:
        unique_signal, output_value = string.split(' | ')
        preprocess.append([unique_signal.split(' '), output_value.split(' ')])
    return preprocess
```

```python
input_list = preprocess_input(read_file())

count = 0

for line in input_list:
    for digit in line[1]:
        if len(digit) == 2:
            # digit = 1
            count += 1
        elif len(digit) == 4:
            # digit = 4
            count += 1
        elif len(digit) == 3:
            # digit = 7
            count += 1
        elif len(digit) == 7:
            # digit = 8
            count += 1

print(f"Instances of digits that use a unique number of segments: {count}")
```

```python
d = {
    "0": set([char for char in "abcefg"]), # 6 letters
    "1": set([char for char in "cf"]),     # 2 letters 
    "2": set([char for char in "acdeg"]),  # 5 letters
    "3": set([char for char in "acdfg"]),  # 5 letters
    "4": set([char for char in "bcdf"]),   # 4 letters
    "5": set([char for char in "abdfg"]),  # 5 letters
    "6": set([char for char in "abdefg"]), # 6 letters
    "7": set([char for char in "acf"]),    # 3 letters
    "8": set([char for char in "abcdefg"]),# 7 letters
    "9": set([char for char in "abcdfg"]), # 6 letters
}

d_list = [
    set([char for char in "abcefg"]),
    set([char for char in "cf"]),
    set([char for char in "acdeg"]),
    set([char for char in "acdfg"]),
    set([char for char in "bcdf"]),
    set([char for char in "abdfg"]),
    set([char for char in "abdefg"]),
    set([char for char in "acf"]),
    set([char for char in "abcdefg"]),
    set([char for char in "abcdfg"]),
]

all_letters = set(char for char in "abcdefg")

def map_easy_digits(signal):
    letter_map = {letter: set(letter for letter in all_letters) for letter in all_letters}
    # Easy words
    for word in signal:
        if len(word) == 2:
            # digit = 1
            for letter in word:
                letter_map[letter] = letter_map[letter].intersection(d["1"])
            
            for letter in all_letters:
                if letter not in word:
                    letter_map[letter] = letter_map[letter].difference(d["1"])
            
            
        elif len(word) == 4:
            # digit = 4
            for letter in word:
                letter_map[letter] = letter_map[letter].intersection(d["4"])
                
            for letter in all_letters:
                if letter not in word:
                    letter_map[letter] = letter_map[letter].difference(d["4"])
            
                    
        elif len(word) == 3:
            # digit = 7
            for letter in word:
                letter_map[letter] = letter_map[letter].intersection(d["7"])
                
            for letter in all_letters:
                if letter not in word:
                    letter_map[letter] = letter_map[letter].difference(d["7"])
                    
            
        elif len(word) == 7:
            # digit = 8
            for letter in word:
                letter_map[letter] = letter_map[letter].intersection(d["8"])
                
            for letter in all_letters:
                if letter not in word:
                    letter_map[letter] = letter_map[letter].difference(d["8"])
                    
    
    return letter_map

def check_mapping_completed(letter_map):
    for letter in letter_map:
        if len(letter_map[letter]) != 1:
            return False
    return True

def update_found_letter_map(letter_map):
    for letter in letter_map:
        if len(letter_map[letter]) == 1:
            for letter2 in letter_map:
                if letter2 != letter:
                    letter_map[letter2] = letter_map[letter2].difference(letter_map[letter])

def new_word_map(letter_map, signal):
    word_map = {}
    for word in signal:
        ls = [list(letter_map[letter]) for letter in word]
        
        combination_words = search_all_possible_words_by_letters(ls, "")
        duplicated_list = [set([char for char in word]) for word in combination_words]
        word_map[word] = ""
        for item in duplicated_list:
            if item != word_map[word] and item in list(d.values()):
                word_map[word] = item
                break
    return word_map
            
def search_all_possible_words_by_letters(list_of_list_of_letters, word):
    # UN BUEN DFS
    if len(list_of_list_of_letters) == 1:
        return [word+char for char in list_of_list_of_letters[0] if char not in word]
    
    ls = []
    for letter in list_of_list_of_letters[0]:
        if letter not in word:
            ls += search_all_possible_words_by_letters(list_of_list_of_letters[1:], word+letter)
    return ls

def deduce_map_digits(signal):
    letter_map = map_easy_digits(signal)
    # Now signal only has words with 5 or 6 letters
    update_found_letter_map(letter_map)
    word_map = new_word_map(letter_map, signal)
    
    while not check_mapping_completed(letter_map):
        for letter in letter_map:
            if len(letter_map[letter]) != 1:
                words = [word for word in signal if letter in word]
                intersection = letter_map[letter]
                for word in words:
                    intersection = intersection.intersection(word_map[word])
                letter_map[letter] = intersection
        
        update_found_letter_map(letter_map)
        
    return {letter: list(letter_map[letter])[0] for letter in letter_map}
```

```python
input_list = preprocess_input(read_file())
final_sum = 0
for item in input_list:
    signal = item[0]
    letter_map = deduce_map_digits(signal)
    
    number = ""
    for word in item[1]:
        ls = set([letter_map[letter] for letter in word])
        i = d_list.index(ls)
        number += str(i)
    
    final_sum += int(number)
    
print(f"Sum of all numbers is {final_sum}")
```
