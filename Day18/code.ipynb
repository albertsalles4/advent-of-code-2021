```python
import os
import numpy as np

def read_file(filename="input.txt", path=os.path.join(os.getcwd())):
    with open(os.path.join(path, filename), "r") as f:
        input_list = [line.strip() for line in f]
    return input_list
```

```python
def preprocess_input(input_list):
    return [eval(line) for line in input_list]

def compute_magnitude(snailfish_number):
    
    if type(snailfish_number) == int:
        return snailfish_number
    
    return 3*compute_magnitude(snailfish_number[0])+2*compute_magnitude(snailfish_number[1])

def explode(sn_str, start_pos):
    end_pos = start_pos+4
    while sn_str[end_pos] != ']':
        end_pos += 1
    
    pair = eval(sn_str[start_pos:end_pos+1])
    
    left_side = sn_str[:start_pos]
    i = len(left_side)-1
    while i >= 0:
        if left_side[i].isnumeric():
            j = i-1
            while left_side[j].isnumeric():
                j -= 1
            num = int(left_side[j+1:i+1])
            left_side = left_side[:j+1] + str(num+pair[0]) + left_side[i+1:]
            break
        i -= 1
        
    right_side = sn_str[end_pos+1:]
    i = 0
    while i < len(right_side):
        if right_side[i].isnumeric():
            j = i+1
            while right_side[j].isnumeric():
                j += 1
            num = int(right_side[i:j])
            right_side = right_side[:i] + str(num+pair[1]) + right_side[j:]
            break
        i += 1
    
    return left_side + "0" + right_side

def find_explosion(sn):
    sn_str = str(sn)
    
    depth = 0
    for i in range(len(sn_str)):
        if sn_str[i] == '[':
            depth += 1
        elif sn_str[i] == ']':
            depth -= 1
        
        if depth == 5:
            return eval(explode(sn_str, i)), True
        
        i += 1
    return sn, False

def find_split(sn):
    sn_str = str(sn)
    
    for i in range(len(sn_str)):
        if sn_str[i].isnumeric():
            j = i+1
            while sn_str[j].isnumeric():
                j += 1
            if j > i+1:
                num = int(sn_str[i:j])
                new_pair = [num//2, num-num//2]
                return eval(sn_str[:i] + str(new_pair) + sn_str[j:]), True 
            
        previous_was_numeric = sn_str[i].isnumeric()
    return sn, False

def reduce_sn(sn):
    explode = True
    split = True
    
    while explode or split:
        sn, explode = find_explosion(sn)
        if not explode:
            sn, split = find_split(sn)
        
    return sn
```

```python
input_list = preprocess_input(read_file())

sn = input_list[0]
for i in range(1, len(input_list)):
    sn = reduce_sn([sn, input_list[i]])
print(f"Magnitude of result is {compute_magnitude(sn)}")
```

```python
input_list = preprocess_input(read_file())

m = 0
for i in range(len(input_list)):
    sn1 = input_list[i]
    for j in range(len(input_list)):
        if i != j:
            sn2 = input_list[j]
            sn = reduce_sn([sn1, sn2])
            magnitude = compute_magnitude(sn)
            m = max(magnitude, m)
print(f"Maximum magnitude adding two numbers is {m}")
```
